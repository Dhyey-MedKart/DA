```python
task_2 = pd.DataFrame(columns=["store_id","Total_sales","Gen_bill_amount","Branded_bill_amount","OTC_bill_amount","Null_billing_amount"])
```

```python
for row in storewise_sales_withcat.values:
    cat = row[4]
    if cat in ['Branded Medicine','Branded medicine','Branded Direct Medicine','Branded Direct medicine','Branded Multivitamin']:
        task_2.loc[len(task_2)] = [row[0],row[2],0,row[2],0,0]
    elif cat in ['Generic Medicine','Generic direct medicine','Generic medicine','Generic Multivitamin','Generic']:
        task_2.loc[len(task_2)] = [row[0],row[2],row[2],0,0,0]
    elif cat in ['Generic OTC', 'Branded OTC','Private OTC']:
        task_2.loc[len(task_2)] = [row[0],row[2],0,0,row[2],0]
    else:
        task_2.loc[len(task_2)] = [row[0],row[2],0,0,0,row[2]]
```

```python
task_2 = task_2.groupby(['store_id'])[['Total_sales',"Gen_bill_amount","Branded_bill_amount","OTC_bill_amount","Null_billing_amount"]].sum()
task_2 = task_2.reset_index()
```

```python
task_2 = task_2.merge(stores,how='left',left_on='store_id',right_on='id')
```
